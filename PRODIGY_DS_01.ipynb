```python
df=pd.read_csv('cities_r2.csv')
```

```python
city_grouped = df.groupby('name_of_city')[['male_graduates', 'female_graduates']].sum()

city_grouped['total_graduates'] = city_grouped['male_graduates'] + city_grouped['female_graduates']

city_sorted = city_grouped.sort_values(by='total_graduates', ascending=False)

top_10_cities = city_sorted.head(10)

cities = top_10_cities.index
indices = np.arange(len(cities))

male_graduates = top_10_cities['male_graduates']
female_graduates = top_10_cities['female_graduates']

bar_width = 0.35

plt.bar(indices, male_graduates, bar_width, label='Male Graduates')
plt.bar(indices + bar_width, female_graduates, bar_width, label='Female Graduates')

plt.xlabel('City')
plt.ylabel('Number of Graduates')
plt.title('Number of Male and Female Graduates for Top 10 Cities')
plt.xticks(indices + bar_width / 2, cities)
plt.legend()

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
